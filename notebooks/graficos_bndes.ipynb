```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_filtrado = pd.read_csv('../base_dados/dados_filtrados.csv')
```

```python
#Fazendo primeiros testes de gráficos

plt.figure(figsize=(12, 8))
sns.set(style = "whitegrid")

#Gerando o gráfico
sns.barplot(x = 'empresa', y = 'valor_recebido_por_empresa', hue = 'numero_srm', data = df_filtrado)

#Personalizando o gráfico
plt.title('Valor Recebido por Cada Empresa no Consórcio')
plt.xlabel('Empresa')
plt.ylabel('Valor Recebido (R$)')
plt.xticks(rotation = 90)
plt.legend(title = "Número SRM")
plt.show()
```

```python

plt.figure(figsize=(18, 10))
sns.set(style = "whitegrid")

#Gerando o gráfico
grafico = sns.barplot(x = 'empresa', y = 'valor_recebido_por_empresa', hue = 'numero_srm', data = df_viva_ufrj, palette = 'pastel')

#Colocando os rótulos
for p in grafico.patches:
    grafico.annotate(f'{p.get_height():,.2f}', 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha = 'center', va = 'center', 
                     fontsize = 12, color = 'black', 
                     xytext = (0, 5), textcoords = 'offset points')

#Personalizando o gráfico
plt.title('Valor Recebido por Cada Empresa no Consórcio')
plt.xlabel('Empresa', color = 'blue')
plt.ylabel('Valor Recebido (R$)', color = 'blue')
plt.xticks(rotation = 0)
plt.legend(title = "Número SRM")
plt.show()
```
